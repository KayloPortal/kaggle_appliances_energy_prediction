```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("./KAG_energydata_complete.csv")
target = data.iloc[:, 1].values
rv1 = data.iloc[:, -2].values
rv2 = data.iloc[:, -1].values
```

```python
correlation_data = pd.DataFrame({'target': target, 'rv1': rv1})

# Calculate the correlation matrix
correlation_matrix = correlation_data.corr()

# Set up the matplotlib figure
plt.figure(figsize=(8, 6))

# Create a heatmap using seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar_kws={"shrink": .8})

# Add titles and labels
plt.title('Correlation Heatmap between rv1 and Target')
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Show the plot
plt.show()
```

```python
correlation_data = pd.DataFrame({'target': target, 'rv2': rv2})

# Calculate the correlation matrix
correlation_matrix = correlation_data.corr()

# Set up the matplotlib figure
plt.figure(figsize=(8, 6))

# Create a heatmap using seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar_kws={"shrink": .8})

# Add titles and labels
plt.title('Correlation Heatmap between rv2 and Target')
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Show the plot
plt.show()
```
